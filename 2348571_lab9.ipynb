```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def unit_step(x, threshold=0):
    return np.where(x >= threshold, 1, 0)

def sign_function(x):
    return np.sign(x)

def linear_function(x):
    return x

def piecewise_linear(x):
    return np.piecewise(x, [x < 0, x >= 0], [lambda x: 0, lambda x: x])

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

```

```python

x_values = np.linspace(-10, 10, 1000)

unit_step_output = unit_step(x_values, threshold=5)
sign_function_output = sign_function(x_values)
linear_function_output = linear_function(x_values)
piecewise_linear_output = piecewise_linear(x_values)
sigmoid_output = sigmoid(x_values)
tanh_output = tanh(x_values)
```

```python


plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.plot(x_values, unit_step_output, label='Unit Step',color='blue')
plt.title('Unit Step Activation Function')
plt.legend()

plt.subplot(3, 2, 2)
plt.plot(x_values, sign_function_output, label='Sign',color='black')
plt.title('Sign Activation Function')
plt.legend()

plt.subplot(3, 2, 3)
plt.plot(x_values, linear_function_output, label='Linear',color='yellow')
plt.title('Linear Activation Function')
plt.legend()

plt.subplot(3, 2, 4)
plt.plot(x_values, piecewise_linear_output, label='Piece-wise Linear',color='red')
plt.title('Piece-wise Linear Activation Function')
plt.legend()

plt.subplot(3, 2, 5)
plt.plot(x_values, sigmoid_output, label='Sigmoid',color='green')
plt.title('Logistic (Sigmoid) Activation Function')
plt.legend()

plt.subplot(3, 2, 6)
plt.plot(x_values, tanh_output, label='Hyperbolic Tangent',color='purplem')
plt.title('Hyperbolic Tangent Activation Function')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
sepal_length = iris.data[:, 0]
```

```python

unit_step_output = unit_step(sepal_length, threshold=5)
sign_function_output = sign_function(sepal_length)
linear_function_output = linear_function(sepal_length)
piecewise_linear_output = piecewise_linear(sepal_length)
sigmoid_output = sigmoid(sepal_length)
tanh_output = tanh(sepal_length)
```

```python

plt.figure(figsize=(12, 10))

plt.subplot(3, 2, 1)
plt.scatter(sepal_length, unit_step_output, label='Unit Step', color='blue')
plt.title('Unit Step Activation Function')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Output')
plt.legend()

plt.subplot(3, 2, 2)
plt.scatter(sepal_length, sign_function_output, label='Sign', color='green')
plt.title('Sign Activation Function')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Output')
plt.legend()

plt.subplot(3, 2, 3)
plt.scatter(sepal_length, linear_function_output, label='Linear', color='red')
plt.title('Linear Activation Function')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Output')
plt.legend()

plt.subplot(3, 2, 4)
plt.scatter(sepal_length, piecewise_linear_output, label='Piece-wise Linear', color='purple')
plt.title('Piece-wise Linear Activation Function')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Output')
plt.legend()

plt.subplot(3, 2, 5)
plt.scatter(sepal_length, sigmoid_output, label='Sigmoid', color='orange')
plt.title('Logistic (Sigmoid) Activation Function')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Output')
plt.legend()

plt.subplot(3, 2, 6)
plt.scatter(sepal_length, tanh_output, label='Hyperbolic Tangent', color='brown')
plt.title('Hyperbolic Tangent Activation Function')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Output')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def perceptron_output(x, y, w1, w2, b):
    xout = x * w1 + b
    yout = y * w2 + b


    and_result = 1 if (xout >= 0) and (yout >= 0) else 0
    and_error = abs(and_result - (x and y))


    or_result = 1 if (xout >= 0) or (yout >= 0) else 0
    or_error = abs(or_result - (x or y))

    xor_result = int(xout >= 0) ^ int(yout >= 0)
    xor_error = abs(xor_result - (x != y))

    return and_error, or_error, xor_error
```

```python
x = int(input("Enter 0 or 1: "))
y = int(input("Enter 0 or 1: "))


w1 = 0.5
w2 = 0.243
b = -0.2
```

```python

and_error, or_error, xor_error = perceptron_output(x, y, w1, w2, b)

print("Error for AND operation:", and_error)
print("Error for OR operation:", or_error)
print("Error for XOR operation:", xor_error)
```
